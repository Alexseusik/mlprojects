```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv("HealthExpend.csv")
```

```python
data
```

```python
X = data[["AGE", "famsize", "COUNTIP", "COUNTOP", "EXPENDIP"]]
Y = data["EXPENDOP"]
```

```python
plt.scatter(X["AGE"], Y, color='red')
plt.title('AGE vs EXPENDOP', fontsize=14)
plt.xlabel('AGE', fontsize=14)
plt.ylabel('EXPENDOP', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(X["famsize"], Y, color='red')
plt.title('famsize vs EXPENDOP', fontsize=14)
plt.xlabel('famsize', fontsize=14)
plt.ylabel('EXPENDOP', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(X["COUNTIP"], Y, color='red')
plt.title('COUNTIP vs EXPENDOP', fontsize=14)
plt.xlabel('COUNTIP', fontsize=14)
plt.ylabel('EXPENDOP', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(X["EXPENDIP"], Y, color='red')
plt.title('EXPENDIP vs EXPENDOP', fontsize=14)
plt.xlabel('EXPENDIP', fontsize=14)
plt.ylabel('EXPENDOP', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(X["COUNTOP"], Y, color='red')
plt.title('COUNTOP vs EXPENDOP', fontsize=14)
plt.xlabel('COUNTOP', fontsize=14)
plt.ylabel('EXPENDOP', fontsize=14)
plt.grid(True)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, 
    test_size = 0.2, 
    random_state = 42
)
```

```python
x_train = sm.add_constant(x_train)
x_test = sm.add_constant(x_test)
```

```python
ols_model = sm.OLS(y_train, x_train).fit()
```

```python
y_pred_train = ols_model.predict(x_train)
y_pred_test = ols_model.predict(x_test)
```

```python
print(ols_model.summary())
```

```python
metrics.r2_score(y_test, y_pred_test)
```
